```python
import pandas as pd
```

```python
song = pd.read_csv('../../../소프트웨어응용/데이터/Spotify_song/tracks_features.csv', encoding = 'utf8')
song.head()
```

```python
song['artist_ids'] = song['artist_ids'].str.split("'").str[1]
song['artists'] = song['artists'].str.split("'").str[1]
song.head()
```

```python
song_attr = song[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                'instrumentalness', 'liveness', 'valence', 'tempo']]
song_attr.head()
```

```python
song_attr.corr()
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns
import plotly.express as px

song2 = song.copy()
features = song_attr.columns

tsne = TSNE(n_components=2).fit_transform(song2.loc[:10000, features])
tsne_df = pd.DataFrame({'x': tsne[:, 0], 'y':tsne[:, 1]})

# fig = px.scatter(
#     tsne_df,
#     x = 'x',
#     y = 'y')

sns.scatterplot(
    x = 'x', y = 'y',
    palette = sns.color_palette("Set1", 10),
    data = tsne_df,
    legend = "full",
    alpha = 0.4
)

plt.title("tSNE")

plt.show()
```

```python
import plotly.express as px
from sklearn.decomposition import PCA

pca = PCA(n_components=3, random_state=42)

features = song_attr.columns

song2 = song.copy()
song2 = song2.sample(frac = 1)
song2 = song2.sample(frac = 1).reset_index(drop = True)

components = pca.fit_transform(song2.loc[:2000, features])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(3)
)

fig.update_traces(diagonal_visible=False)
fig.show()
```
